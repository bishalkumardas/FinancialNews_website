```python
# Define ticker symbol and expiration date
ticker_symbol = "AAPL"
expiration_date = "2025-02-14"

# Fetch stock data
stock = yf.Ticker(ticker_symbol)
options = stock.option_chain(expiration_date)

# Define relevant columns
columns_to_keep = ["lastTradeDate", "contractSymbol", "strike", "lastPrice", "bid", "ask", "openInterest", "volume"]

# Function to process option data (calls & puts)
def process_options(data):
    df = pd.DataFrame(data)[columns_to_keep].copy()
    df.insert(0, "tradeDate", df["lastTradeDate"].dt.strftime("%d %b %Y"))  # Insert at first position
    df.insert(1, "tradeTime", df["lastTradeDate"].dt.strftime("%I:%M %p"))  # Insert at second position
    return df.drop(columns=["lastTradeDate"])  # Drop old date column

# Process Calls & Puts
calls_df = process_options(options.calls)
puts_df = process_options(options.puts)

# Display the processed data
print("Calls:\n", calls_df.head(), "\n")
print("Puts:\n", puts_df.head())
```

```python
import pandas as pd
import requests
from io import StringIO

# Define the URL
csv_url = "https://nsearchives.nseindia.com/content/equities/EQUITY_L.csv"

# Define headers with a User-Agent
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36"
}

# Make the request
response = requests.get(csv_url, headers=headers)

# Check if the request was successful
if response.status_code == 200:
    # Read the CSV into a DataFrame
    df = pd.read_csv(StringIO(response.text))
    # print(df.head())  # Display first few rows
    df = df[["SYMBOL", "NAME OF COMPANY"]]

    df['SYMBOL']= df["SYMBOL"].astype(str) + ".NS"
    df
else:
    print(f"Failed to fetch data. Status code: {response.status_code}")
```
